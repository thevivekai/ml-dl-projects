# Right no. of Cluster using ELbow Method

```python
X = data.values
```

```python
X
```

```python
# within cluster sum of squares . it's a metric used in clustering (espect k means clsutering) to compactness of the clsuters
wcss = []
K = range(1,11)
```

```python
for k in K:
    kmeans = KMeans(n_clusters=k,random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

# Plot the Elbow curve

```python
plt.figure(figsize=(8,6))
plt.plot(K,wcss,'bo-')
plt.xlabel('number of clusters(k)')
plt.ylabel('wcss')
plt.title('Elbow Method for Optimal K')
```
