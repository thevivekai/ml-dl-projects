## Simple Linear Regression (With 1 Independent Variable(x) & 1 Dependent Variable(y))

### Step 1: Import the Dataset

```python
import pandas as pd
data = pd.read_csv("Boston1.csv")
data
```

```python
data.shape
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

### Step 2: Figuring Out Independent(x) & dependent(y) Variable

##### first we take the 1 as a independent variable and 1 as a dependent variable which are last two colomns

```python
data_ = data.loc[:,['lstat','medv']]
data_.head()
```

#### Visualizing variables

```python
import matplotlib.pyplot as plt
data_.plot(x='lstat',y='medv',style='o')
plt.xlabel('lstat')
plt.ylabel('medv')
plt.show() 
```

#### Preparing Data

```python
x = pd.DataFrame(data_['lstat'])
y = pd.DataFrame(data_['medv'])
```

```python
x.size,y.size
```

### Step 3: Split the dataset into x_train, x_test, y_train & y_test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(x_test.shape)
```

### Step 4: Creating the Model using Scikit Learn Library

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

### Step 5: Train the Model

```python
regressor.fit(x_train,y_train)
```

```python
print(regressor.intercept_)   # it basically the 'c' in 'y=mx+c'
```

```python
print(regressor.coef_)  # it basically the 'm' in 'y=mx+c'
```

### Step 6: Predicting using the testing dataset

```python
y_pred = regressor.predict(x_test)
y_pred = pd.DataFrame(y_pred,columns=['Predicted'])
```

```python
y_pred
```

```python
y_test
```

### Step 7: Evaluate the performance of Model

```python
from sklearn import metrics
import numpy as np
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_pred))
print('Mean Squared Error:',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```

```python
# r^2 value 
r2 = regressor.score(x_test, y_test)
print(f"R-squared value: {r2:.4f}")
```

## Multiple Linear Regression (With More than 1 Independent Variable(x) & 1 Dependent Variable(y))

### Step 1: Import the Dataset

```python
dataset = pd.read_csv("Boston1.csv")
dataset
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset.describe()
```

### Step 2: Figuring Out Independent(x) & dependent(y) Variables

```python
X = pd.DataFrame(dataset.iloc[:,:-1])
y = pd.DataFrame(dataset['medv'])
```

```python
X.shape
```

```python
y.shape
```

### Step 3: Split the dataset into x_train, x_test, y_train & y_test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Step 4: Creating the Model using Scikit Learn Library

```python
from sklearn.linear_model import LinearRegression
mulreg = LinearRegression()
```

### Step 5: Train the Model

```python
mulreg.fit(X_train,y_train)
```

```python
m = pd.DataFrame(mulreg.coef_,index=['Coeffienct']).transpose()
a = pd.DataFrame(X.columns,columns=['Attribute'])
```

```python
m
```

```python
a
```

```python
coef = pd.concat([m,a],axis=1,join='inner')
coef
```

```python
print(mulreg.intercept_)
```

### Step 6: Predicting using the testing dataset

```python
y_pred = mulreg.predict(X_test)
y_pred = pd.DataFrame(y_pred,columns=['Predicted'])
```

```python
y_pred
```

```python
y_test
```

### Step 7: Evaluate the performance of Model

```python
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_pred))
print('Mean Squared Error:',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```

```python
# r^2 value 
r2 = mulreg.score(X_test, y_test)
print(f"R-squared value: {r2:.4f}")
```
