# Random Forest Classifier Demonstration

### Import the Libraries & Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
iris.target
```

```python
iris.data
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
iris_data = pd.DataFrame({'sepal length':iris.data[:,0],
                  'sepal width':iris.data[:,1],
                  'petal length':iris.data[:,2],
                  'petal width':iris.data[:,3],
                  'species':iris.target})
```

### Find x & y

```python
x = iris_data.iloc[:,0:4]
y = iris_data.iloc[:,4]
```

```python
x
```

```python
y
```

### Split the Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

### Model Creation

```python
from sklearn.ensemble import RandomForestClassifier
ran_clf = RandomForestClassifier(n_estimators=100)  # n_estimator  represent the how many number of tree we want is it given by programmer
ran_clf.fit(x_train,y_train)
```

### Visualization of Random Forest Tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plot_tree(ran_clf.estimators_[0],filled=True)
plt.show()
```

### Visualize with out Graph

```python
from sklearn.tree import export_text
tree_rules = export_text(ran_clf.estimators_[0],feature_names = list(x_train.columns))
print(tree_rules)
```

### For All Tree without graph

```python
for idx,tree in enumerate(ran_clf.estimators_):
    print(f"Tree{idx}")
    print(export_text(tree,feature_names = list(x_train.columns)))
```

### Model Prediction

```python
y_pred = ran_clf.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

### Model Evaluation

```python
from sklearn.metrics import confusion_matrix,accuracy_score
result = confusion_matrix(y_test,y_pred)
print(result)
```

```python
result2 = accuracy_score(y_test,y_pred)
print(result2)
```
